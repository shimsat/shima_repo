```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page_source)
```

```python
#Web Scraping
#soup.find('span', class_='temp').get_text()
```

```python
temps = []
for tag in soup.find_all('span', class_="wxperiod_temp"):
    temps.append(tag.get_text())
    
```

```python
print(temps)
```

```python
import pandas as pd
temps_series = pd.Series(temps)
temps_series
```

```python
time = []
for tag in soup.find_all('span', class_="date"):
    time.append(tag.get_text())
print(time)
```

```python
time_series = pd.Series(time)
time_series
```

```python
day = []
for tag in soup.find_all('span', class_="day"):
    day.append(tag.get_text())
print(day)
```

```python
day_series = pd.Series(day)
day_series
```

```python
cloud_cover = []
for tag in soup.find_all('span', class_="wx_description"):
    cloud_cover.append(tag.get_text())
cloud_cover
```

```python
cloud_series = pd.Series(cloud_cover)
cloud_series
```

```python
wind = []
for tag in soup.find_all('span', class_="wxObs"):
    wind.append(tag.get_text())
wind
```

```python
#Cleaning Data

wind2 = ['0' if (x == '-' or x == '<1' ) else x for x in wind]
wind2
```

```python
wind_series = pd.Series(wind2)
wind_series.head(72)
```

```python
frame = {'temps': temps_series, 'wind':wind_series.head(72), 'cloud_cover':cloud_series, 'time': time_series, 'day': day_series}
df = pd.DataFrame(frame)
df
```

```python
import datetime
current_time = datetime.datetime.now()#.strftime('%Y-%m-%d %H:%M:%S')
```

```python

number_of_hours = len(time)

date_list = [(current_time + datetime.timedelta(hours = hour)) for hour in range(number_of_hours)] 
date_list
```

```python
df['datetime'] = pd.Series(date_list)
df['datetime'] = pd.to_datetime(df.datetime)
```

```python
df.temps = pd.to_numeric(df.temps)
df.wind = pd.to_numeric(df.wind)
```

```python
df.dtypes
#df = df.set_index(['datetime'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.drop(['time', 'day'], axis=1)
```

```python
df.dtypes
```

```python
df.set_index(['datetime'], inplace=True)
```

```python
#Temperature
#plt.plot(df.temps)
df.temps.plot()
plt.xticks(rotation=45)
plt.ylabel('degree C')
```

```python
df['temp_class'] = df['temps'].apply(lambda x: 'mild' if (x<25 and x>18) else ('cold' if x>13 and x <18 else 'very cold'))
df
```

```python
dfplot_temp= df.groupby(['temp_class']).size().reset_index(name='counts')
dfplot_temp
```

```python
labels = 'cold', 'mild', 'very cold'
sizes = dfplot_temp['counts']
#colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0.1, 0.1)  # explode 1st slice

# Plot
plt.pie(sizes,explode=explode, labels=labels,
autopct='%4.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
